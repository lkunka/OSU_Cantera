```python
# Python
#Solve a 0D simulation for detailed chemical kinetics using Cantera
import cantera as ct
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import copy
import sdtoolbox
import yt
```

```python
gas_0D = ct.Solution('h2o2.cti')
fuel_species="H2"
ox_species="O2"
phi=1
gas_0D.set_equivalence_ratio(phi,fuel_species,ox_species)
gas_0D.TP=1100, 101325
```

```python
r = ct.Reactor(contents=gas_0D)
reactorNetwork=ct.ReactorNet([r])
```

```python
t=0.0
dt=1.0E-9
est_time=0.00009
counter=1
time=[t]
state=[]
X=[gas_0D.X]
Temp=[gas_0D.T]
q=[0.0]

while(t < est_time):
    t=t+dt
    reactorNetwork.advance(t)
    if (counter%1 == 0):
        time.append(t)
        X.append(gas_0D.X)
        Temp.append(gas_0D.T)
        state.append(reactorNetwork.get_state())
        # Note that the species concentrations are mass fractions
        #timeHistory.loc[t] = reactorNetwork.get_state()
        q.append(-np.sum(gas_0D.net_production_rates*gas_0D.partial_molar_enthalpies))
    counter+=1
print(counter)
time=np.array(time)
state=np.array(state)
X=np.array(X)
Temp=np.array(Temp)
```

```python
fig,ax1=plt.subplots(figsize=(10,10))
for i in range (3,reactorNetwork.n_vars):
    ax1.plot(time, X[:,i-3],'--',label=reactorNetwork.component_name(i).split()[1])
ax1.tick_params(axis='y', labelcolor='k')

ax2=ax1.twinx()
ax2.plot(time,Temp,'r')
ax2.tick_params(axis='y', labelcolor='r')

ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Species Mass Fraction", color='k')
ax2.set_ylabel("Temperature (K)", color='r')
ax1.legend(loc="upper left", fontsize="small")
plt.title("Detailed Kinetics")
plt.savefig("0D_H2O2_Sim.png")
plt.show()

print("Final Temp: ", Temp[-1])
#print("Final Density: ", cpp_data[-1,2])
#print("Final Pressure: ", cpp_data[-1,3])
```

```python

filePath="/Users/logankunka/0D_FAST/working_dir/plot_files_methaneDDT/"
ts=yt.DatasetSeries(filePath+"plt*")
yt.funcs.mylog.setLevel(50)

fast_time=[]
fast_temp=[]
yF=[]
yO=[]
yP=[]

for ds in ts.piter():
        fast_time.append(ds.current_time)
        fast_temp.append(ds.find_max('temperature')[0])
        yF.append(ds.find_max('fuel mass fraction')[0])
        yP.append(ds.find_max('product mass fractio')[0])
        yO.append(ds.find_max('air mass fraction')[0])
fast_time=np.array(fast_time)
fast_temp=np.array(fast_temp)
yF=np.array(yF)
yP=np.array(yP)
yO=np.array(yO)
```

```python
fig,ax1=plt.subplots(figsize=(10,10))
ax1.plot(fast_time, yF, '--', label="Fuel Mass Fraction")
ax1.plot(fast_time, yP, '--',label="Product Mass Fraction")
ax1.plot(fast_time, yO, '--',label="Oxidizer Mass Fraction")
ax1.tick_params(axis='y', labelcolor='k')

fast_temp_scaled=fast_temp-(fast_temp-fast_temp[0])/(fast_temp[-1]-fast_temp[0])*fast_temp[0]
#scaling_factor=
ax2=ax1.twinx()
ax2.plot(fast_time,fast_temp,'r',label="Temperature")
ax2.plot(fast_time,fast_temp_scaled,'r--',label="Adjusted Temperature")
ax2.tick_params(axis='y', labelcolor='r')
                     
ax1.set_xlabel("Time (s)")
ax2.set_ylabel("Temperature (K)", color='r')
ax1.set_ylabel("Mass Fraction", color='k')
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
plt.title("Tuned CDM Kinetics")
plt.savefig("0D_H2O2_CDM.png")
plt.show()
```
